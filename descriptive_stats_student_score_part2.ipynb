```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('student-scores.csv')

df.head(5)
```

```python
cols_subjects = ['math_score', 'history_score', 'physics_score', 'chemistry_score','biology_score', 'english_score', 'geography_score']
q1 = df[cols_subjects].quantile(0.25).tolist()
q2 = df[cols_subjects].quantile(0.5).tolist()
q3 = df[cols_subjects].quantile(0.75).tolist()

i = 0
for subject in cols_subjects:
    print(f"{str(cols_subjects[i])} {q1[i]}, {q2[i]}, {q3[i]}")
    i+=1

```

```python
import matplotlib.pyplot as plt

avg_scores = df[cols_subjects].mean(axis=1)
q1, q2, q3 = avg_scores.quantile([0.25, 0.5, 0.75]).tolist()
plt.hist(avg_scores, bins=10, color='blue', edgecolor='white')
plt.axvline(q1, color='green', linestyle='--', label='Q1(25th %)')
plt.axvline(q2, color='red', linestyle='--', label='Median(50th %)')
plt.axvline(q3, color='blue', linestyle='--', label='Q3(75th %)')

plt.title("Subjects distribution with Quartiles")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
stdev_math_scores = df[cols_subjects[0]].std()
avg_math_scores = df[cols_subjects[0]].mean()
df_z_score = (df[cols_subjects] - avg_math_scores) / stdev_math_scores
df_z_score
```
